```python
import pandas as pd
import numpy as np

df = pd.read_csv("ci_main.csv")
df.head()
```

```python
df["actual"] = df["captcha_image"].apply(lambda x: x.split("/")[-1].split(".")[0])
df.head()
```

```python
df.drop(columns=["collection_id", "captcha_image", "mode"], inplace=True)
df.head()
```

```python
df["is_correct"] = np.where(df["actual"] == df["response"], 1, 0)
df.head()
```

```python
df = df[[col for col in df.columns if col != 'response'] + ['response']]
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
df.shape
```

```python
df.to_csv("ci_main_cleaned.csv", index=False)
df.head()
```

```python
import ast
df["mouse_movements"] = df["mouse_movements"].apply(ast.literal_eval)
df = df[df["mouse_movements"].apply(lambda x: len(x) > 10)].reset_index(drop=True)
```

```python
df.to_csv("ci_main_cleaned.csv", index=False)
df.head()
```
